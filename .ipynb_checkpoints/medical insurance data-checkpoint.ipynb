### medical insurance data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor ,plot_tree
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
df= pd.read_csv('medical_insurance.csv')
df
```

```python
df.info()
```

```python
sns.boxplot(df['age'])
```

```python
sns.boxenplot(df['bmi'])
```

```python
sns.boxenplot(df['charges'])
```

### eda part

```python
df['gender'].value_counts().to_dict()
```

```python
df['gender'].replace({'male': 0, 'female': 1},inplace=True)
df
```

```python
df['smoker'].value_counts().to_dict()
```

```python
df['smoker'].replace({'no': 0, 'yes': 1},inplace=True)
df
```

```python
df=pd.get_dummies(df,['region'],dtype=int)
```

```python
df
```

### model training

```python
x= df.drop(['charges'],axis=1)
y=df['charges']
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
dt_reg=DecisionTreeRegressor(random_state=1)
dt_reg.fit(x_train,y_train)
```

```python
dt_reg.feature_importances_
```

### model evaluation 

```python
y_predict_train=dt_reg.predict(x_train)
```

### Trainng data set accuracy

```python
mse=mean_absolute_error(y_train,y_predict_train)
print('mean_absolute_error',mse)
rmse= np.sqrt(mse)
print('rmse',rmse)
r2score=r2_score(y_train,y_predict_train)
print('r2score',r2score)
```

### testing data set accuracy 

```python
y_predict_test=dt_reg.predict(x_test)
```

```python
mse=mean_absolute_error(y_test,y_predict_test)
print('mean_absolute_error',mse)
rmse= np.sqrt(mse)
print('rmse',rmse)
r2score=r2_score(y_test,y_predict_test)
print('r2score',r2score)
```

```python
plt.figure(figsize=(100,80))
plot_tree(dt_reg)
plt.savefig('tree.jpg')
```
